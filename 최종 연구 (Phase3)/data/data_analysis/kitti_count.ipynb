## Bounding Box

### bbox 면적 계산

```python
# bbox의 면적을 label[15]에 추가
label[15] = (label[6]-label[4])*(label[7]-label[5])
label
```

```python
image_size = 1242*375
label['bbox'] = label[15] / image_size
label
```

### 범주 별 개수 구하기

```python
bins = [0, 0.0001, 0.001, 0.01, 0.1, 1]
labels = ['~0.01%','0.01%~0.1%', '0.1%~1%', '1%~10%', '10%~']
label['ratio'] = pd.cut(label['bbox'], bins, labels=labels)
label
```

```python
# bbox 면적 별 객체 수
label['ratio'].value_counts()
```

```python
BoundingBox = pd.value_counts(label['ratio'].values, sort=False)
BoundingBox.plot.bar()
BoundingBox.plot.bar(figsize=(10,8), fontsize=15)
```
